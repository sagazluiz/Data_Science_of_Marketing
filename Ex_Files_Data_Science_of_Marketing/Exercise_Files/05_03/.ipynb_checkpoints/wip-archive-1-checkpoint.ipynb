```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
 
import pandas as pd
import numpy as np
 
# Only needed if you want to display your plots inline if using Notebook
# change inline to auto if you have Spyder installed
%matplotlib inline
```

```python
# import some data to play with
# iris = datasets.load_iris()
# ^ thats the original 

myClusterData = pd.read_csv('~/Documents/devel/dsom/6-cluster/data/cluster_python.csv', index_col=0)
```

```python
# assuming we're assigning this data to a data frame using pd and calling that dataframe x
# x = pd.DataFrame(iris.data)
# ^ thats the original 

x = pd.DataFrame(myClusterData)

myClusterData.head(3)

# myClusterData.head(3)

# to see some of what we've got here 
# x.head(3)

# to see all of what we've got here
# x
```

```python
# This separates out the inputs 
# x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
# ^ thats the original 

x.columns = ['b2','b3','bf','cta','age','geo','sex']
```

```python
#y = pd.DataFrame(iris.target)
# ^ thats the original 

y = pd.DataFrame(myClusterData.b3)
```

```python
# y.columns = ['Targets']
# ^ thats the original 

y.columns = ['Targets']
```

```python
plt.figure(figsize=(14,7))
```

```python
colormap = np.array(['red', 'lime', 'black'])
```

```python
# Plot Sepal
# plt.subplot(1, 2, 1)
# plt.scatter(x.Sepal_Length, x.Sepal_Width, c=colormap[y.Targets], s=40)
# plt.title('Sepal')

plt.subplot(1,2,1)
plt.scatter(x.cta, x.sex, c=colormap[y.Targets], s=40)

# da, this doesn't work - yes and - i think its a data issue - look at the data whosit is using here and model mine after 
```

```python
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Petal')
```

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
model.labels_
```

```python
# View the results
# Set the size of the plot
plt.figure(figsize=(14,7))
 
# Create a colormap
colormap = np.array(['red', 'lime', 'black'])
 
# Plot the Original Classifications
plt.subplot(1, 2, 1)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
 
# Plot the Models Classifications
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('K Mean Classification')
```

```python
# The fix, we convert all the 1s to 0s and 0s to 1s.
predY = np.choose(model.labels_, [1, 0, 2]).astype(np.int64)
print (model.labels_)
print (predY)
```

```python
# View the results
# Set the size of the plot
plt.figure(figsize=(14,7))
 
# Create a colormap
colormap = np.array(['red', 'lime', 'black'])
 
# Plot Orginal
plt.subplot(1, 2, 1)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
 
# Plot Predicted with corrected values
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[predY], s=40)
plt.title('K Mean Classification')
```

```python
# Performance Metrics
sm.accuracy_score(y, predY)
```

```python
# Confusion Matrix
sm.confusion_matrix(y, predY)
```
